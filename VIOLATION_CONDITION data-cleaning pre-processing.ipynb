```python
import pandas as pd
import numpy as np
import random
```

```python
# Load the dataset into a Pandas DataFrame
df_if = pd.read_csv("C:/Users/vrl2k/Desktop/MS/Summer 2023/Data Cleaning/Project/INSPECTION_FACT.csv")
```

```python
VIOLATION_COLUMNS = ['INSPECTION_ID','INSPECTION_DATE','INSPECTION_TYPE','VIOLATION_CONDTION','RISK_TYPE','INSPECTION_RESULT']
df_vc1 = df_if[VIOLATION_COLUMNS]
```

```python
df_vc1.dtypes
```

```python
# Convert the date format to '%m/%d/%Y'
df_vc1['INSPECTION_DATE'].fillna(0, inplace=True)
df_vc1['INSPECTION_DATE'] = pd.to_datetime(df_vc1['INSPECTION_DATE'], format='%Y-%m-%dT%H:%M:%SZ')
```

```python
# Split the 'delimited_field' into a list of values and create a new DataFrame with the split values
df_vc_split = df_vc1['VIOLATION_CONDTION'].str.split('|', expand=True)
df_vc_split
```

```python
# Concatenate the split DataFrame with the original DataFrame
df_vc2 = pd.concat([df_vc1, df_vc_split], axis=1)
```

```python
# Remove VIOLATION_CONDTION column
column_to_drop = 'VIOLATION_CONDTION'
df_vc2.drop(column_to_drop, axis=1, inplace=True)
```

```python
# Rename the columns 
df_vc2.rename(columns={0: 'VIOLATION_CONDTION_1', 1: 'VIOLATION_CONDTION_2', 2: 'VIOLATION_CONDTION_3', 3: 'VIOLATION_CONDTION_4', 4: 'VIOLATION_CONDTION_5', 5: 'VIOLATION_CONDTION_6', 6: 'VIOLATION_CONDTION_7', 7: 'VIOLATION_CONDTION_8', 8: 'VIOLATION_CONDTION_9', 9: 'VIOLATION_CONDTION_10', 10: 'VIOLATION_CONDTION_11', 11: 'VIOLATION_CONDTION_12', 12: 'VIOLATION_CONDTION_13', 13: 'VIOLATION_CONDTION_14', 14: 'VIOLATION_CONDTION_15', 15: 'VIOLATION_CONDTION_16', 16: 'VIOLATION_CONDTION_17', 17: 'VIOLATION_CONDTION_18', 18: 'VIOLATION_CONDTION_19', 19: 'VIOLATION_CONDTION_20', 20: 'VIOLATION_CONDTION_21', 21: 'VIOLATION_CONDTION_22', 22: 'VIOLATION_CONDTION_23', 23: 'VIOLATION_CONDTION_24'}, inplace=True)
df_vc2
```
